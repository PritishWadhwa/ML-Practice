# Gradient Descent: In General

```python
import numpy as np
import matplotlib.pyplot as plt
```

hardcoded X and Y functions

```python
X = np.arange(10)
Y = (X-5)**2
```

```python
print(X, Y)
```

### Goal: Given a function f(x), we want to find value of x that minimizes f

Visualize the function

```python
plt.style.use('seaborn')
plt.plot(X, Y)
plt.ylabel('f(x)')
plt.xlabel('x')
plt.show()
```

```python
x = 0
# step size = learning rate(lr)
lr = 0.1
# f(x) denotes error 
error = []
# take 50 steps in downhill direction
plt.plot(X, Y)
for i in range(50):
    grad = 2 * (x - 5)
    x = x - lr*grad
    y = (x-5)**2
    error.append(y)
    plt.scatter(x, y)
    print(x)
```

plot the values of error

```python
plt.plot(error)
plt.show()
```
