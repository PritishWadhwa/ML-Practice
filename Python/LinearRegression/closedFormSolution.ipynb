```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import make_regression
```

```python
# Generate Dataset
X, y = make_regression(n_samples=400, n_features=1, n_informative=1, noise=1.8, random_state=11)
print(X.shape, y.shape)
```

```python
y = y.reshape((-1, 1))
print(X.shape, y.shape)
```

```python
# Normalize
u = X.mean()
std = X.std()
X = (X-u)/std
```

```python
# Visualize Dataset
plt.style.use('seaborn')
plt.figure()
plt.scatter(X, y)
plt.show()
```

```python
ones = np.ones((X.shape[0], 1))
# X_ = np.hstack((ones, X))
X_ = np.hstack((X, ones))
print(X_.shape)
```

```python
def predict(X, theta):
    return np.dot(X, theta)
```

```python
# have to compute (X'X)-1X'y
def getThetaUsingClosedForm(X, y):
    # convert y to a matrix from a vector
    y = np.mat(y)
    firstPart = np.dot(X.T, X)
    secondPart = np.dot(X.T, y)
    theta = np.linalg.pinv(firstPart)*secondPart
    return theta
```

```python
theta = getThetaUsingClosedForm(X_, y)
theta
```

```python
plt.close()
plt.figure()
plt.scatter(X, y)
plt.plot(X, predict(X_, theta), color='red')
plt.show()
```
