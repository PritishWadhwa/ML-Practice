```python
X = X[:, 1:]
Y = Y[:, 1:].reshape((-1,))
```

```python
queryX = np.array([0,3]) #queryPoint
plt.scatter(x=X[:, 0], y=X[:, 1], c=Y, cmap='autumn')
plt.scatter(queryX[0], queryX[1], color='blue')
plt.show()
```

```python
def dist(x1, x2):
    ans = np.sqrt(sum((x1-x2)**2)) #calculates euclidian distance
    return ans
```

```python
#function to give knn
def knn(X, Y, queryPoint, k = 5):
    vals = []
    m = X.shape[0]
    for i in range(m):
        d = dist(queryPoint, X[i])
        vals.append((d, Y[i]))
    vals = sorted(vals)
    vals = vals[:k]
    vals = np.array(vals)
    newVals = np.unique(vals[:, 1], return_counts = True)
    maxFreqIndex = newVals[1].argmax()
    prediction = newVals[0][maxFreqIndex] 
    return prediction
```

```python
x = knn(X, Y, queryX)
print(x)
```
