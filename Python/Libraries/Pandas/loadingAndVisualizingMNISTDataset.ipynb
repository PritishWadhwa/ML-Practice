```python
data = df.values
np.random.shuffle(data)
```

```python
print(data)
print(type(data))
print(data.shape)
```

```python
X = data[:, 1:] #pixel values for numbers
Y = data[:, 0]  #numper corresponding to the data
```

```python
print(X.shape)
print(Y.shape)
```

```python
X[0].shape
```

```python
#To visualize this, I have to convert it to a 2D array of size 28 x 28
def drawImg(X, Y, i):
    plt.imshow(X[i].reshape(28, 28), cmap='gray')
    plt.title("Label " + str(Y[i]))
    plt.show()
```

```python
for i in range(10):
    drawImg(X, Y, i)
```

```python
#split the training set
split = int(0.8*X.shape[0])
print(split)
```

```python
X_train, Y_train = X[:split, :], Y[:split]
X_test, Y_test = X[split:, :], Y[split:]
print(X_train.shape, Y_train.shape)
print(X_test.shape, Y_test.shape)
```

```python
#alternate way to do the above
XT, Xt, YT, Yt = train_test_split(X, Y, test_size= 0.2, random_state=1)
#if random state not included, it will always give different value
print(XT.shape, YT.shape)
print(Xt.shape, Yt.shape)
```

```python
#plot a visualization of (grid of first 25 images 5x5 grid)
plt.figure(figsize=(10, 10))
for i in range(25):
    plt.subplot(5, 5, i+1)
    plt.imshow(X_train[i].reshape(28, 28), cmap='gray')
    plt.axis('off')
    plt.title(Y_train[i])
```
