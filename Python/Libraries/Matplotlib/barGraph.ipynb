```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.bar([0, 1, 2], [10, 20, 15]) #[x axis][bar lenghts]
plt.show()
```

```python
xCoordinates = np.array([0, 1, 2])*2
plt.bar(xCoordinates - 0.25, [10, 20, 15], width = 0.5, label = 'current year', tick_label = ['gold', 'platinum', 'silver'])
plt.bar(xCoordinates + 0.25, [20, 10, 12], width = 0.5, label ='next year')
plt.legend()
plt.title("Metal price comparison")
plt.xlabel('metal')
plt.ylabel('price')
plt.show()
```

```python
plt.style.use("dark_background")
xCoordinates = np.array([0, 1, 2])*2
plt.bar(xCoordinates - 0.25, [10, 20, 15], width = 0.5, label = 'current year', tick_label = ['gold', 'platinum', 'silver'], color = 'green')
plt.bar(xCoordinates + 0.25, [20, 10, 12], width = 0.5, label ='next year', color = 'orange')
plt.legend()
plt.title("Metal price comparison")
plt.xlabel('metal')
plt.ylabel('price')
plt.ylim(0, 30)
plt.xlim(-1, 5)
plt.show()
```
