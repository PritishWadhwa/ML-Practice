```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use("seaborn")
subject = "Maths", "Chem", "Phy", "Comp Science" 
weightage = [20, 30, 10, 15]
plt.pie(weightage, labels=subject, explode=(0.2, 0, 0, 0), autopct='%1.1f%%', shadow=True)
plt.title("subjects")
plt.show()
```
