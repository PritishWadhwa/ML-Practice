```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Xsn = np.random.randn(100)
u = 70
sigma = 8
X = Xsn * sigma + u
```

```python
print(X)
```

```python
X = np.round(X)
print(X)
```

```python
plt.style.use("seaborn")
plt.hist(X, rwidth=0.95)
plt.ylabel('frequency of students')
plt.xlabel('Marks Range')
plt.show()
```

```python
X2 = np.round(Xsn*8 + 40)
plt.hist(X, rwidth=0.95, alpha=0.8, label='Maths')
plt.hist(X2, rwidth=0.90, alpha=0.8, label='Physics')
plt.ylabel('frequency of students')
plt.xlabel('Marks Range')
plt.legend()
plt.title("Histogram")
plt.show()
```
