```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Line Plot
x = np.arange(10)
y1 = x**2
y2 = 2*x + 3
print(x)
print(y1)
print(y2)
```

```python
plt.plot(x, y1)
plt.show()
```

```python
themes = plt.style.available
print(themes)
```

```python
plt.style.use("seaborn")
```

```python
plt.plot(x, y1, color = 'red', label = 'apple', marker = 'o')
plt.plot(x, y2, color = 'green', label = 'kiwi', linestyle = 'dashed', marker = '*', markersize = 12)
plt.xlabel('time')
plt.ylabel('price')
plt.title("Price of fruits over time")
plt.legend()
plt.show()
```
