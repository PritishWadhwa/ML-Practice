```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
prices = np.array([1,2,3,4])**2
print(prices)
```

```python
plt.plot(prices) #if given only array, it becomes y axis and array indices become x axis
plt.show()
```

```python
x = np.arange(10)
y1 = x**2
y2 = 2*x + 3
```

```python
plt.scatter(x, y1)
plt.scatter(x, y2)
plt.show()
```

```python
#adjust size of plot
plt.figure(figsize=(10,4))
plt.scatter(x, y1, color = 'red', label = 'apple', marker = '^')
plt.scatter(x, y2, color = 'green', label = 'kiwi', linestyle = 'dashed', marker = '*')
plt.xlabel('time')
plt.ylabel('price')
plt.title("Price of fruits over time")
plt.legend()
plt.show()
```
